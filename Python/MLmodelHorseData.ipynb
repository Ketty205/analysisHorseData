```python
#load library for works with dataset
import pandas as pd
```

```python
data1 = pd.read_csv('HorseFinall.csv', delimiter=',', encoding='latin-1')
data1
```

```python
#created outcome variable, that definding the class with each observation. Yes-for the 1st final place, No for the rest places
```

```python
print(type(data1['Final.place']))
```

```python
import numpy as np
```

```python
# adding column y
data1['y'] = pd.Series(np.zeros(data1.shape[0]))
# adding value based on condition
data1.loc[data1['Prize.money']<1.450000e+06,'y'] = 'Yes'
data1.loc[data1['Prize.money']>=1.450000e+06,'y'] = 'No'
```

```python
data1.head()
```

```python
#First ml model to be maked is Knn-works with numerical and categorical data
#Columns to be used: Track, Surface, Country, RaceType, Distance, Starting.position, Jockey.weight, Horse.age, Race.time, Final.place
```

```python
#Preprocessing data- mapping values of categorical columns 
```

```python
data1['Track'].unique()
```

```python
data1['Track'] = data1['Track'].map({'Sha Tin':0,'Happy Valley':1})
```

```python
data1['Surface'].unique()
```

```python
data1['Surface'] = data1['Surface'].map({'Gress':0,'Dirt':1})
```

```python
data1['RaceType'].unique()
```

```python
data1['RaceType'] = data1['RaceType'].map({'Handicap':0,'Non-handicap':1,'Griffinl?p':2})
```

```python
data1['Country'].unique()
```

```python
data1['Country'] = data1['Country'].map({'Sverige':0,'Danmark':1, 'Irland':2, 'Australia':3,'USA':4,'New Zealand':5,'Great Britain':6,'Japan':7,'Frankrike':8,'Tyskland':9,'Italia':10,'Spania':11,'S?r-Afrika':12,'Argentina':13,'Brasil':14,'Uten Land':15,'Chile':16})
```

```python
X1 = data1[['Track','Surface', 'Country', 'RaceType', 'Distance','Starting.position', 'Jockey.weight', 'Horse.age', 'Race.time','Final.place']]
```

```python
data1['y'] = data1['y'].map({'No':0,'Yes':1})
```

```python
data1.isnull()
```

```python
y1 = data1[['y']]
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train1)
X_train1 = scaler.transform(X_train1)
X_test1 = scaler.transform(X_test1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train1,np.ravel(y_train1))
prediction = knn.predict(X_test1)
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,recall_score, f1_score
print(confusion_matrix(y_test1, prediction))
print("Accuracy:",accuracy_score(y_test1, prediction)*100)
print("Precision:",precision_score(y_test1, prediction,average='weighted',labels=np.unique(prediction))*100)
print("Recall:",recall_score(y_test1, prediction,average='weighted')*100)
print("F1:",f1_score(y_test1, prediction,average='weighted')*100)
```

```python
#Optimizing KNN
error = []
for i in  range(1,100):
    knn = KNeighborsClassifier(n_neighbors= i,metric='euclidean')
    knn.fit(X_train1,np.ravel(y_train1))
    pred_i = knn.predict(X_test1)
    error.append(np.mean(pred_i != np.ravel(y_test1)))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.plot(range(1,100),error)
plt.title('K-values')
plt.xlabel('K')
plt.ylabel('Error')
```

```python
#Optimizing KNN
error = []
for i in  range(1,100):
    knn = KNeighborsClassifier(n_neighbors= i,metric='manhattan')
    knn.fit(X_train1,np.ravel(y_train1))
    pred_i = knn.predict(X_test1)
    error.append(np.mean(pred_i != np.ravel(y_test1)))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,100),error)
plt.title('K-values')
plt.xlabel('K')
plt.ylabel('Error')
```

```python
from sklearn.model_selection import GridSearchCV
grid_params = {
    'n_neighbors': range(1,100),
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan',]
}
gs = GridSearchCV(
    KNeighborsClassifier(),
    grid_params,
    verbose=3,
    cv=3,
    scoring='accuracy'
)
gs_results = gs.fit(X_train1, np.ravel(y_train1))  
```

```python
gs_results.best_params_
```
